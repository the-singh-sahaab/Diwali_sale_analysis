```python
ax = sns.countplot(x = "Gender", data = df)
for bar in ax.containers:
    ax.bar_label(bar)
```

```python
sale_gen = df.groupby(['Gender'], as_index = False)['Amount'].sum().sort_values(by = 'Amount', ascending = False)
sns.barplot(x = 'Gender', y = "Amount", data = sale_gen)
```

from above graph we got this, the most number of sale was done to the female buyes and most buyers were female only.

```python
ax = sns.countplot(data = df, x = 'Age Group', hue = "Gender")
for i in ax.containers:
    ax.bar_label(i)
```

```python
sale_gen = df.groupby(['Age Group'], as_index = False)['Amount'].sum().sort_values(by = 'Amount', ascending = False)
sns.barplot(x = 'Age Group', y = "Amount", data = sale_gen)
```

```python
sale_state = df.groupby(['State'], as_index = False)["Orders"].sum().sort_values(by = 'Orders', ascending = False).head(10)
sns.set(rc = {'figure.figsize': (15, 5)})
sns.barplot(data = sale_state, x = 'State', y = 'Orders')
```

from above we conclued which get and which state.


```python
#marital state
ax = sns.countplot(data = df, x = 'Marital_Status')
for i in ax.containers:
    ax.bar_label(i)
#concluding how marital public bought more product
```

```python
#Occupation
a = sns.countplot(data = df, x = 'Occupation')
for i in ax.containers:
    ax.bar_label(i)
# showing of what occupation people where there to but 
```

```python
sale_state = df.groupby(['Occupation'], as_index = False)['Amount'].sum().sort_values(by = 'Amount', ascending = False)

# sns.set(rc = {'figure.figsaze': (20, 5)})
sns.barplot(data = sale_state, x = 'Occupation', y = 'Amount')
```

we got from above that which sector spend the most on different product.


```python
#product analysis

sns.set(rc= {'figure.figsize':(20,5)})
ax = sns.countplot(data = df, x = "Product_Category")
```

```python
sale_product = df.groupby(["Product_ID"],  as_index = False)['Orders'].sum().sort_values(by = 'Orders', ascending = False).head(10)
sns.barplot(data = sale_product, x = 'Product_ID', y = "Orders")
```

```python
sale_product = df.groupby(["Product_Category"],  as_index = False)['Amount'].sum().sort_values(by = 'Amount', ascending = False).head(10)
sns.barplot(data = sale_product, x = 'Product_Category', y = "Amount")
```
